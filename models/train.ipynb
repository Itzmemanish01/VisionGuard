```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')

model.train(
    data=r'../data/raw/no-helmet.v1i.yolov8/data.yaml',
    epochs=30,
    imgsz=640,
    batch=8,
    name='visionguard_helmet'
)
```

```python
model = YOLO(r'../models/runs/detect/visionguard_helmet-4/weights/best.pt')
results = model(r'../data/sample_images/image4.png')
results[0].show()
```
